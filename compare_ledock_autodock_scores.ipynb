```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
adata = np.genfromtxt('a_score_table.csv', dtype=[('name', 'S100'), ('score', 'f')])
adata.sort(order='name')
```

```python
rdata = np.genfromtxt('r_score_table.csv', dtype=[('name', 'S100'), ('score', 'f')])
rdata.sort(order='name')
```

```python
%matplotlib inline
f, ax = plt.subplots(figsize=(6, 6))
ax.scatter(adata['score'], rdata['score'])
ax.plot(ax.get_xlim(), ax.get_ylim(), ls="--", c=".3")
plt.xlabel('atom')
plt.ylabel('residue')
```
